```python
from tensorflow.keras.models import load_model
import joblib
import numpy as np

model=load_model('model.h5')
scaler=joblib.load('scaler.pkl')
type_encoder=joblib.load('type_encoder.pkl')
```

```python
def predict_fraud(step, type_str, amount, oldbalanceOrg, newbalanceOrig, oldbalanceDest, newbalanceDest):

    type_encoded=type_encoder.transform([type_str])[0]

    numerical_features = np.array([[
            step,
            amount,
            oldbalanceOrg,
            newbalanceOrig,
            oldbalanceDest,
            newbalanceDest
        ]])
    numerical_scaled=scaler.transform(numerical_features)
    model_input = np.array([[
        numerical_scaled[0][0],   # step
        type_encoded,             # type (encoded)
        numerical_scaled[0][1],   # amount
        numerical_scaled[0][2],   # oldbalanceOrg
        numerical_scaled[0][3],   # newbalanceOrig
        numerical_scaled[0][4],   # oldbalanceDest
        numerical_scaled[0][5]    # newbalanceDest
    ]])
    probal=model.predict(model_input)[0][0]

    prediction = int(probal >= 0.5)

    return probal, prediction

result,pred=predict_fraud(step=4,
    type_str="TRANSFER",
    amount=850000,
    oldbalanceOrg=900000,
    newbalanceOrig=0,
    oldbalanceDest=1000,
    newbalanceDest=851000)

print("fraud_probability :",result)
print("prediction :",pred)
```
